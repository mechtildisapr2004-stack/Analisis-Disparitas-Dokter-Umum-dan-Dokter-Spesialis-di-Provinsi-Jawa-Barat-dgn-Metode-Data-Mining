10. Menggabungkan kedua dataset dalam satu DataFrame

```python
df_gabungan = pd.merge(df_umum, df_spesialis, on=['nama_kabupaten_kota', 'tahun', 'kode_kabupaten_kota', 'kode_provinsi', 'nama_provinsi', 'satuan'], how='inner')

# Mengatur ulang urutan kolom
cols = df_gabungan.columns.tolist()
# Pindahkan 'tahun' ke posisi sebelum 'jumlah_dokter_umum'
# Temukan indeks kolom 'tahun' dan 'jumlah_dokter_umum'
tahun_index = cols.index('tahun')
umum_index = cols.index('jumlah_dokter_umum')

if tahun_index > umum_index:
    cols.pop(tahun_index)
    cols.insert(umum_index, 'tahun')

df_gabungan = df_gabungan[cols]

# Tampilkan hasil penggabungan dalam bentuk tabel
print("\nDataFrame Gabungan:")
display(df_gabungan.head())
print(f"Total Baris Setelah Digabung: {len(df_gabungan)}")
```

11. Menghitung Rata-Rata Disparitas per Tahun

```python
average_disparity_per_year = df_gabungan.groupby('tahun')['disparitas_dokter'].mean().reset_index()

# Tampilkan hasilnya
print("\nRata-rata Disparitas Dokter per Tahun:")
display(average_disparity_per_year)
```

12. Menghitung perbedaan mutlak antara jumlah dokter umum dan spesialis

```python
df_gabungan['disparitas_dokter'] = abs(df_gabungan['jumlah_dokter_umum'] - df_gabungan['jumlah_dokter_spesialis'])

# Tampilkan dataframe dengan kolom disparitas baru
print("\nDataFrame Gabungan dengan Kolom Disparitas:")
display(df_gabungan.head())
```

13. Visualisasi Rata-Rata Disparitas Dokter per Tahun

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Atur gaya visualisasi
sns.set_style("whitegrid")

# Buat plot garis
plt.figure(figsize=(10, 6))
sns.lineplot(data=average_disparity_per_year, x='tahun', y='disparitas_dokter', marker='o')

# Tambahkan judul dan label
plt.title('Rata-rata Disparitas Dokter per Tahun di Jawa Barat', fontsize=14)
plt.xlabel('Tahun', fontsize=12)
plt.ylabel('Rata-rata Disparitas Dokter', fontsize=12)
plt.xticks(average_disparity_per_year['tahun']) # Menampilkan semua tahun di sumbu x
plt.grid(True)

# Tampilkan plot
plt.tight_layout()
plt.show()
```

14. Menghitung 10 Kabupaten/Kota dengan Rata-rata Disparitas Dokter Tertinggi (2016-2024)

```python
# Hitung rata-rata disparitas per kabupaten/kota selama semua tahun
average_disparity_per_kabkota = df_gabungan.groupby('nama_kabupaten_kota')['disparitas_dokter'].mean().reset_index()

# Urutkan hasilnya dari yang tertinggi ke terendah dan ambil 10 teratas
top_10_disparity_kabkota = average_disparity_per_kabkota.sort_values(by='disparitas_dokter', ascending=False).head(10)

# Tampilkan 10 kabupaten/kota dengan rata-rata disparitas tertinggi
print("\n10 Kabupaten/Kota dengan Rata-rata Disparitas Dokter Tertinggi (2016-2024):")
display(top_10_disparity_kabkota)

# Buat visualisasi bar plot untuk 10 teratas
plt.figure(figsize=(12, 7))
sns.barplot(data=top_10_disparity_kabkota, x='disparitas_dokter', y='nama_kabupaten_kota', palette='viridis')

plt.title('10 Kabupaten/Kota dengan Rata-rata Disparitas Dokter Tertinggi (2016-2024)', fontsize=14)
plt.xlabel('Rata-rata Disparitas Dokter', fontsize=12)
plt.ylabel('Nama Kabupaten/Kota', fontsize=12)
plt.tight_layout()
plt.show()
```

15. Menghitung 10 Kabupaten/Kota dengan Rata-rata Disparitas Dokter Terendah (2016-2024)

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Urutkan hasilnya dari yang terendah ke tertinggi dan ambil 10 terendah
bottom_10_disparity_kabkota = average_disparity_per_kabkota.sort_values(by='disparitas_dokter', ascending=True).head(10)

# Tampilkan 10 kabupaten/kota dengan rata-rata disparitas terendah
print("\n10 Kabupaten/Kota dengan Rata-rata Disparitas Dokter Terendah (2016-2024):")
display(bottom_10_disparity_kabkota)

# Buat visualisasi bar plot untuk 10 terendah
plt.figure(figsize=(12, 7))
sns.barplot(data=bottom_10_disparity_kabkota, x='disparitas_dokter', y='nama_kabupaten_kota', palette='viridis')

plt.title('10 Kabupaten/Kota dengan Rata-rata Disparitas Dokter Terendah (2016-2024)', fontsize=14)
plt.xlabel('Rata-rata Disparitas Dokter', fontsize=12)
plt.ylabel('Nama Kabupaten/Kota', fontsize=12)
plt.tight_layout()
plt.show()
```

16. Menghitung Data untuk Plot Tren Rasio dan Menampilkan Tren Rasio Dokter Umum vs Spesialis di Kab/Kota Kunci

```python
df_gabungan['Rasio_DU_DS'] = df_gabungan['jumlah_dokter_umum'] / (df_gabungan['jumlah_dokter_spesialis'] + 1e-6)

# Pilih beberapa Kab/Kota kunci untuk perbandingan tren (sesuai contoh user, gunakan nama kolom yang benar)
kota_kunci = ['KOTA BANDUNG', 'KABUPATEN GARUT', 'KABUPATEN BEKASI', 'KABUPATEN KUNINGAN']
df_tren_rasio = df_gabungan[df_gabungan['nama_kabupaten_kota'].isin(kota_kunci)].copy()

# Tampilkan data yang akan diplot
print("Data untuk Plot Tren Rasio:")
display(df_tren_rasio.head())

# Buat visualisasi line plot
plt.figure(figsize=(14, 7))
sns.lineplot(
    x='tahun',
    y='Rasio_DU_DS',  # Rasio Dokter Umum : Spesialis
    hue='nama_kabupaten_kota',
    data=df_tren_rasio,
    marker='o'
)

plt.title('Tren Rasio Dokter Umum vs Spesialis di Kab/Kota Kunci (2016-2024)', fontsize=14)
plt.xlabel('Tahun', fontsize=12)
plt.ylabel('Rasio Dokter Umum / Dokter Spesialis', fontsize=12)
plt.xticks(df_tren_rasio['tahun'].unique())
plt.grid(True)
plt.legend(title='Kabupaten/Kota')
plt.tight_layout()
plt.show()
```

17. Visualisasi Metode Elbow untuk Menentukan Jumlah Klaster Optimal

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# Pilih fitur untuk klasterisasi
features = ['jumlah_dokter_umum', 'jumlah_dokter_spesialis', 'Rasio_DU_DS']
X = df_gabungan[features].copy()

# Scaling fitur
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Hitung inertia untuk berbagai jumlah klaster (K)
inertia = []
k_range = range(1, 11) # Coba K dari 1 hingga 10

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10) # n_init=10 untuk menghindari warning
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)

# Plot hasil Metode Elbow
plt.figure(figsize=(10, 6))
plt.plot(k_range, inertia, marker='o')
plt.title('Metode Elbow untuk Menentukan Jumlah Klaster Optimal')
plt.xlabel('Jumlah Klaster (K)')
plt.ylabel('Inertia')
plt.xticks(k_range)
plt.grid(True)
plt.show()
```

18. DataFrame Gabungan dengan Label Klaster

```python
# Lakukan Klasterisasi K-Means dengan jumlah klaster = 3
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
df_gabungan['cluster_label'] = kmeans.fit_predict(X_scaled)

# Tampilkan beberapa baris pertama dengan label klaster
print("\nDataFrame Gabungan dengan Label Klaster:")
display(df_gabungan.head())
```

19. Menghitung Rata-rata Fitur per Klaster

```python
# Analisis karakteristik klaster dengan menghitung rata-rata fitur untuk setiap klaster
cluster_characteristics = df_gabungan.groupby('cluster_label')[features].mean()

# Tampilkan hasil analisis klaster
print("\nRata-rata Fitur per Klaster:")
display(cluster_characteristics)
```

```python
# Menggunakan data asli df_gabungan dengan label klaster baru
sns.pairplot(df_gabungan, vars=features, hue='cluster_label', palette='viridis')
plt.suptitle('Visualisasi Klaster K-Means (K=3)', y=1.02) # Tambahkan judul di atas
plt.show()
```
